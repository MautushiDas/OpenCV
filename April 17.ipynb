```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
import os
from skimage.morphology import convex_hull_image
import pandas as pd
```

```python
img = cv.imread('./star.png',0) # reading image in greysacle
#img = cv.imread('./flower.png',0) # reading image in greysacle
window_name = 'image' 
plt.imshow(img)
# cv.imshow(window_name,img)
```

```python
#img = cv.imread('.\star.png',0) # reading image in greysacle
ret,thresh = cv.threshold(img,127,255,0) # theresohd fuction retruns the new values of pixel based on thresold value
print(ret)
print('thesh',thresh)
contours,hierarchy = cv.findContours(thresh, 1, 2)
cnt = contours[0]
#print(cnt.shape)
M = cv.moments(cnt)
```

```python
cx = int(M['m10']/M['m00'])
cy = int(M['m01']/M['m00'])
area = cv.contourArea(cnt)
perimeter = cv.arcLength(cnt,True)
)
```

```python
epsilon = 0.1*cv.arcLength(cnt,True)
approx = cv.approxPolyDP(cnt,epsilon,True)
print(epsilon)
print(approx)
```

```python
hull = cv.convexHull(cnt)
```

```python
k = cv.isContourConvex(cnt)
```

```python
rect = cv.minAreaRect(cnt)
box = cv.boxPoints(rect)
box = np.int0(box)
```
